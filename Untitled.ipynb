```python
ys=[]
yd={}
for i in range(len(d)):
    x=d['Goals by World Cup'][i].split(',')
    for j in range(len(x)):
        k=x[j].split('in')
        k[0]=int(k[0])
        k[1]=int(k[1])
        if(k[1] in ys):
            yd[k[1]].update({d['Player'][i][1:]:k[0]})
        else:    
            ys.append(k[1])
            yd[k[1]]={d['Player'][i][1:]:k[0]}
```

```python
yd1 = dict(sorted(yd.items()))
```

```python
for i in yd1:
    yd1[i]=dict(sorted(yd1[i].items(), key=lambda x: x[1],reverse=True))
```
